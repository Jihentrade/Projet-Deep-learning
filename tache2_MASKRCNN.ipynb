enregistrer les ensembles de données train, val et test pour les utiliser dans l'entraînement du modèle,on a utiliser la fonction register_coco_instances pour enregistrer les ensembles de données sous forme de datasets COCO.

```python
from detectron2.data.datasets import register_coco_instances
register_coco_instances("train", {}, "/content/drive/MyDrive/jihen-1/train/_annotations.coco.json", "/content/drive/MyDrive/jihen-1/train")
register_coco_instances("val", {}, "/content/drive/MyDrive/jihen-1/valid/_annotations.coco.json", "/content/drive/MyDrive/jihen-1/valid")
register_coco_instances("test", {}, "/content/drive/MyDrive/jihen-1/test/_annotations.coco.json", "/content/drive/MyDrive/jihen-1/test")
```

Visualisation de cinq exemples aléatoires d'entraînement

```python
my_dataset_train_metadata = MetadataCatalog.get("train")
dataset_dicts = DatasetCatalog.get("train")

import random
from detectron2.utils.visualizer import Visualizer

for d in random.sample(dataset_dicts, 5):
    img = cv2.imread(d["file_name"])
    visualizer = Visualizer(img[:, :, ::-1], metadata=my_dataset_train_metadata, scale=0.5)
    vis = visualizer.draw_dataset_dict(d)
    cv2_imshow(vis.get_image()[:, :, ::-1])
```
